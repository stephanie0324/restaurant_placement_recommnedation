```python
df = df.drop_duplicates(subset=['business_id','postal_code']).reset_index(drop=True)
df = df.drop(columns=['index', 'business_id', 'stars_x', 'useful', 'funny', 'cool', 'text',
       'date','address','is_open', 'stars_y','attributes','hours'])
```

```python
# sort by name and review count 
df = df.sort_values(['name', 'review_count'],
              ascending = [True, False]).reset_index(drop=True)
```

```python
# create relevance score
cnt = 0
for i in Counter(df.name):
    tmp = df[df.name==i].reset_index(drop=True)
    tmp['relevance']=''
    score = len(tmp)
    for j in range(len(tmp)):
        tmp['relevance'][j]=score-j
    if cnt ==0 :
        new_df = tmp
        cnt+=1
    new_df = pd.concat([new_df,tmp])
```

```python
postal_code_df = new_df[['postal_code','density', 'entropy', 'competitiveness']]
postal_code_df = postal_code_df.drop_duplicates(subset=['postal_code'])
postal_code_df.reset_index(inplace=True, drop=True)
```

```python
postal_code_info = new_df.groupby(["postal_code"]).agg({"competitiveness":lambda x:list(set(x))})
```

```python
user_2_loc_df = new_df.groupby("name").agg({"postal_code":lambda x:list(set(x))})
user_2_loc_map = dict(zip(user_2_loc_df.index, user_2_loc_df['postal_code']))

candidate_pool = new_df['postal_code'].unique().tolist()
```

```python
def candidate_generation(name, candidate_pool, user_2_loc_map):
    """
    Note: this a totally random generation, only for demo purpose
    Generates a list of N anime candidates for a given user based on their previously liked animes.
    
    Parameters:
        name (int): The user's ID.
        candidate_pool (list): A list of all possible anime candidates.
        user_2_loc_map (dict): A dictionary that maps users to their liked animes.
        N (int): The number of anime candidates to generate.
    
    Returns:
        already_interacted (list): List of animes which user already liked
        candidates (list): A list of N anime candidates for the user.
    """
    
    #get the already liked animes
    already_interacted = user_2_loc_map[name]

    #candidates will be rest of animes which are not exposed to user
    candidates = list(set(candidate_pool) - set(already_interacted))
    
    return already_interacted, candidates
```

```python
def generate_predictions(name,user_2_loc_map,candidate_pool,ranker):
    """
    Generates predictions for anime recommendations for a given user.
    
    Parameters:
        name (int): The user's ID.
        user_2_loc_map (dict): A dictionary that maps users to their liked animes.
        candidate_pool (list): A list of all possible anime candidates.
        feature_columns (list): A list of feature columns to use for generating predictions.
        ranker (object): A trained model object that is used to generate predictions.
        N (int): The number of anime predictions to generate.
        
    Returns:
        predictions (DataFrame): A dataframe containing the top N anime recommendations for the user.
    """
    already_liked,candidates = candidate_generation(name,candidate_pool,user_2_loc_map)
    #Create dataframe for candidates
    candidates_df = pd.DataFrame(data=pd.Series(candidates,name='postal_code'))
    candidates_df = candidates_df.sort_values(by=['postal_code'])
    # Merge with feature dataframe
    features = postal_code_df.merge(candidates_df)
    features = features.sort_values(by=['postal_code'])
    features = features.iloc[:, 1:]
#     print("candidates_df:",candidates_df)
#     print("features:",features)

    # If number of already liked animes is less than number of candidates
    # Extend the already liked list with -1
    
    already_liked = list(already_liked)
    if len(already_liked) < len(candidates):
        append_list = np.full(fill_value=-1,shape=(len(candidates)-len(already_liked)))
        already_liked.extend(list(append_list))
    
    #Create dataframe for predictions
    predictions = pd.DataFrame(index=candidates)
    
    #Generate predictions
    predictions['score'] = ranker.predict(features)
    predictions = predictions.sort_values(by='score',ascending=False)
    return predictions
```

```python
predictions = generate_predictions("Acropolis Greek Taverna",
                                   user_2_loc_map,
                                   candidate_pool,
                                   ranker=model)
```
